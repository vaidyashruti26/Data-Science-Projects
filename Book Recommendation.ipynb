```python
#there will be lot of nan value in our dataframe as it is highly unlikely that ours users
book_users=df.pivot(index='book_idN',columns='user_id',values='rating').fillna(0)
matrix_books_users=csr_matrix(book_users.values)
print(matrix_books_users)
```

```python
book_users.head()
```

```python
#knn with bruteforce algorithm
knn=NearestNeighbors(metric='cosine',algorithm='brute',n_neighbors=10)
```

```python
knn.fit(matrix_books_users)
```

```python
def recommender(book_name,data,model,n_recommendations):
    model.fit(data)
    idx=process.extractOne(book_name,df['original_title'])[2]
    print('Movie Selected:',df['original_title'][idx],'Index:',idx)
    print('Searching for recommendations.....')
    distance, indices=model.kneighbors(data[idx],n_neighbors=n_recommendations)
    for i in indices:
        print(df['original_title'][i].where(i!=idx))
```

```python
recommender('The Great Gatsby',matrix_books_users,knn,5)
```
